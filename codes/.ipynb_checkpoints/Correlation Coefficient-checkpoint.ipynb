```python
def standardizer(x):
    
    """
        standardize entity-to-feature data matrix by 
          applying Z-scoring and Range standardization methods
        
        Arguments: 
            x, numpy array, entity-to-feature data matrix
        
        Returns:
            Z-scored and Range standardized data matrices (x_zscr_std, x_rng_std).
    """
    
    x_ave = np.mean(x, axis=0)
    x_rng = np.ptp(x, axis=0)
    x_std = np.std(x, axis=0)
    x_zscr_std = np.divide(np.subtract(x, x_ave), x_std)   # Z-scoring standardization
    x_rng_std = np.divide(np.subtract(x, x_ave), x_rng)  # Range standardization 
    return x_zscr_std, x_rng_std
```

So we will consider more challenging pair, namely, company abon. and storage features.


- $ a = \frac{\sum_{i=1}^{N} (y_{i} - \bar{y}) . (x_{i} - \bar{x})/N}{\sum_{i=1}^{N}(x_{i} - \bar{x}) (x_{i} - \bar{x})/N} $ 

- $b = \bar{y} − a ·\bar{x}$

- $ \rho = \frac{\sum_{i=1}^{N} (y_{i} - \bar{y}) . (x_{i} - \bar{x})/N}{\sigma(x) \sigma(y)} $

- Moreover, $ a = \rho \frac{\sigma(y)}{\sigma(x)}$

Let us fit Linear Regression then:




```python
# Standardizing Data
# x_zscr_std, x_rng_std = standardizer(x=x_org)  
```

```python
x = x_org[:, 2]
y = x_org[:, 4]
N = x.shape[0]

a = (
    (y-np.mean(y)) @ (x-np.mean(x))/ N) / (
    (x - np.mean(x)) @ (x-np.mean(x))/N)

# rho = ((y-np.mean(y)) @ (x-np.mean(x)))/(np.var(x)*np.var(y))

b = np.mean(y) - a * np.mean(x)
print('y = %.3f' % a, '* x %.3f' % b)
```

- a = 9.935 meaning an increase in the number of company abon. by 1 would increase the storage by 9.935 PetaByte.

```python
# Alternative approach: using SKlearn Lib
# Fitting Linear Regression 
reg = LinearRegression(normalize=False,).fit(x.reshape(-1, 1), y) 
print("slope:", reg.coef_, "intercept: %.3f" % reg.intercept_)
```

```python
fig = plt.figure()
plt.scatter(x, y)
plt.plot(x, a*x + b, color='r')
# plt.plot(x, reg.predict(x.reshape(-1, 1)), color='g')
plt.grid()
plt.xlabel('Company Abon.', size=15)
plt.ylabel('Storage', size=15)
plt.show()
```

- $ \rho = \frac{\sum_{i=1}^{N} (y_{i} - \bar{y}) . (x_{i} - \bar{x})/ N}{\sigma(x) \sigma(y)} $ 


```python
# correlation coefficient
rho = (
    ((y-np.mean(y)) @ (x-np.mean(x))) / N) /(
    np.std(x) * np.std(y))

# determinacy coefficient
rho_2 = np.power(rho, 2)

print("correlation coefficient: %.3f " % rho,  
      "determinacy coefficient %.3f" % rho_2 )
```


- Correlation coefficient is a measure of degree of a linear relation between x and y. 
The closer it is to 1 or -1 the more linearly related are features.

As one can see, the linear relation is not that strong.

- Determinacy coefficient is the proportion of the variance $\sigma^{2}(y)$ taken into account by the linear regression of y over x.

As one can see, the proportion is equal to 29% ( 71% is unexplained)

- Moreover, the slope of regression model is positive which shows a direct relation between two features


### Make prediction on some random points

### Mean Relative Absolute Error

A measure of prediction accuracy of a forecasting method

$mrae = \frac{1}{N} \sum_{i=1}^{N} |\frac{y_{i} - y^{'}_{i}}{y_{i}}|$



Ref: https://en.wikipedia.org/wiki/Mean_absolute_percentage_error


```python
mrae = 100*np.mean(np.absolute(np.divide(y - a*x+b, y)))
print("mean relative absolute error : %.3f" % mrae + "%", "\n"
     "determinacy coefficient:       %.3f" % rho_2)
```


- MRAE  is the prediction accuracy of the model. And in our case it shows 111% error.


### MRAE machine learning style

```python
mrae_ml = 100*np.mean(np.absolute(np.divide(y - a*x+b, a*x+b)))
print("mean relative absolute error : %.3f" % mrae_ml + "%", "\n"
     "determinacy coefficient:       %.3f" % rho_2)
```
